```python
from numpy import random
```

```python
"""
A random number is a number that cannot be predicted logically.
If there is a program to generate random numbers, the random number is not truly random
To generate a true random, we need outside data(keystrokes, mouse movements, network traffic)
"""
```

```python
# Generating random numbers
a = random.randint(100)
b = random.rand() # random floats
print(a)
print(b)

arr = random.randint(100, size=(10)) # array of random numbers
print(arr)

arr1 = random.randint(50, size = (4, 3))
print(arr1)

c = random.rand(4) # generate 4 floats
print(c)

# choice -> generate random value based on array values
arr1 = random.choice([1, 2, 3, 4, 5])
print(arr1)

arr2 = random.choice([1, 2, 3, 5, 9], size=(5, 2))
print(arr2)
```

```python
"""
Normal gaussian distribution - it fits the probability distribution for many events
random.normal() - get normal data distribution
Parameters; - random.normal()
  loc - peak of bell
  scale - standard devivation
  size - shape of returned array
"""
```

```python
# random distribution of size 2 x 3
arr5 = random.normal(size=(2, 3))
print(arr5)

# random normal dist, 2 x 3, mean = 1, std dev = 2
arr6 = random.normal(loc=1, scale=2, size=(2, 3))
print(arr6)
```

```python
# Visualizing normal distribution - bell curve
sb.distplot(random.normal(size=50), hist=False)
```

```python
"""
Binomial distribution - a discrete distribution
Describes the outcome of binary scenarios
Parameters;
  n - trials
  p - probability
  size - shape of returned array
"""
```

```python
arr7 = random.binomial(n=10, p=0.5, size=1000)
# print(arr7)
sb.distplot(arr7, kde=False, hist=True)
```

```python
# Normal vs Binomial distribution
sb.distplot(random.normal(loc=5, scale=5, size=100), hist=False, label="normal")
sb.distplot(random.binomial(n=10, p=0.5, size=100), hist=False, label="binomial")
```

```python
"""
Poisson distribution - a discrete distribution
How many times an event can happen in a specific time frame
Parameters;
  lam -> rate or known occurrences
  size -> shape of returned array
"""
```

```python
# random 1 x 10 distribution for occurrence of 2
arr8 = random.poisson(lam=4, size=1000)
sb.distplot(arr8, kde=False)
```

```python
# Normal and poisson distribution
sb.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label="normal")
sb.distplot(random.poisson(lam=50, size=1000), hist=False, label="poisson")
# plt.show()
```

```python
# Poisson vs Bionomial
sb.distplot(random.poisson(lam=10, size=1000), hist=False, label="poisson")
sb.distplot(random.binomial(n=1000, p=0.1, size=1000), hist=False, label="binomial")
```

```python
"""
Uniform distribution
-> Describes occassions every event has an equal chance of occurrrence
Parameters:
  a -> lower bound
  b -> upper bound
  size -> shape of array returned
"""
```

```python
# 4 x 3 uniform distribution
arr9 = random.uniform(size = (4, 3))
print(arr9)
sb.distplot(arr9, hist=False)
```

```python
sb.distplot(random.uniform(size=1000), hist=False)
```

```python
"""
Logistic distribution - describes growth
Parameters;
  loc -> peak
  scale -> std deviation, distribution flatness
  size -> shape of array returned
"""
```

```python
sb.distplot(random.logistic(size=1000), hist=False)
```

```python
# Logistic vs Normal distribution
sb.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sb.distplot(random.logistic(size=1000), hist=False, label='logistic')
```

```python
"""
Multinomial distribution -> generalizes binomail distribution
Describes outcome of multi-nomial scenarios
Parameters:
  n -> possbible outcomes
  pvals -> probabilities of outcomes
  size -> shape of returned array
"""
```

```python
# A dice throw
outcome = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
print(outcome) # value for each pval
```

```python
"""
Exponential distribution -> describes tim until next event
Parameters:
  scale -> inverse of rate
  size -> shape of array returned
"""
```

```python
# exponential distribution scale=2, 4x3
expo = random.exponential(scale=2, size=(4, 3))
print(expo)
sb.distplot(expo, hist=False)
```

```python
# Poisson vs exponential
sb.distplot(random.exponential(scale=2, size=(4, 3)), hist=False, label="Exponential")
sb.distplot(random.poisson(lam=2, size=1000), hist=False, label="poisson")
```

```python
"""
Chi square distribution -> used to verify a hypothesis
Parameters;
  df = freedom degree
  size = shape of array returned
"""
```

```python
# chi = random.chisquare(df=5, size=(4, 3))
# print(chi)
chi = random.chisquare(df=1, size=1000)
sb.distplot(chi, hist=False)
```

```python
"""
Rayleigh distribution -> used for signal processing
Parameters:
  scale -> standard deviation, flatter deistribution
  size -> ""
"""
```

```python
rayleigh = random.rayleigh(scale=2, size=1000)
sb.distplot(rayleigh, hist=False)
```

```python
# Rayleigh vs chisquare
sb.distplot(random.rayleigh(scale=2, size=1000), hist=False)
sb.distplot(random.chisquare(df=2, size=1000), hist=False)
```
